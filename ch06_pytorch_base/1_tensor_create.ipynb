```python
import torch
import numpy as np
```

6.3.1 基本张量创建

（1）按内容创建张量

```python
# 传入一个标量
tensor1 = torch.tensor(10)
print(tensor1)
print(tensor1.size())
print(tensor1.dtype)
```

```python
# 传入一个一维数组
tensor2 = torch.tensor([1.0, 2, 3])
print(tensor2)
print(tensor2.size())
print(tensor2.dtype)
```

```python
# 传入一个二维ndarray
tensor3 = torch.tensor(np.array([[1.0, 2, 3], [4, 5, 6]]))
print(tensor3)
print(tensor3.size())
print(tensor3.dtype)
```

(2)创建指定形状的张量

```python
tensor1 = torch.Tensor(3, 2, 4) # 底层是 2 * 4 的矩阵，一个 3 个
print(tensor1)
print(tensor1.size())
print(tensor1.dtype)
```

```python
tensor2 = torch.Tensor([[1, 2, 3, 4], [4, 5, 6, 7]])
print(tensor2)
print(tensor2.size())
print(tensor2.dtype)
```

```python
tensor2 = torch.Tensor(np.array([[1, 2, 3, 4], [4, 5, 6, 7]]))
print(tensor2)
print(tensor2.dtype)
```

```python
tensor1 = torch.Tensor(10)
print(tensor1)
print(tensor1.size())
print(tensor1.dtype)
```

(3) 创建指定类型的张量

```python
# int（int32）类型
tensor1 = torch.IntTensor(2, 3)
tensor2 = torch.tensor([1, 2, 3], dtype=torch.int32)
print(tensor1.dtype)
print(tensor2.dtype)
```

```python
# long（int64）类型
tensor1 = torch.LongTensor(2, 3)
tensor2 = torch.tensor([1, 2, 3], dtype=torch.int64)
print(tensor1.dtype)
print(tensor2.dtype)
```

```python
# short（int16）类型
tensor1 = torch.ShortTensor(2, 3)
tensor2 = torch.tensor([1, 2, 3], dtype=torch.int16)
print(tensor1.dtype)
print(tensor2.dtype)
```

```python
tensor1 = torch.ByteTensor(2, 3)
tensor2 = torch.tensor([1, 2, 3], dtype=torch.int8)
print(tensor1.dtype)
print(tensor2.dtype)
```

```python
# float类型
tensor1 = torch.FloatTensor(2, 3)
tensor2 = torch.tensor([1, 2, 3], dtype=torch.float32)
print(tensor1.dtype)
print(tensor2.dtype)
print(tensor1)
```

```python
# double类型
tensor1 = torch.DoubleTensor(2, 3)
tensor2 = torch.tensor([1, 2, 3], dtype=torch.float64)
print(tensor1.dtype)
print(tensor2.dtype)
print(tensor1)
```

```python
# half类型
tensor1 = torch.HalfTensor(2, 3)
tensor2 = torch.tensor([1, 2, 3], dtype=torch.float16)
print(tensor1.dtype)
print(tensor2.dtype)
print(tensor1)
```

```python
# bool类型
tensor1 = torch.BoolTensor(2, 3)
tensor2 = torch.tensor([1, 2, 3], dtype=torch.bool)
print(tensor1.dtype)
print(tensor2.dtype)
print(tensor1)
```

6.3.2 指定区间的张量创建

```python
tensor1 = torch.arange(10, 30, 2)
print(tensor1)
```

```python
tensor2 = torch.arange(6)
print(tensor2)
```

```python
tensor1 = torch.linspace(10, 30, 6)
print(tensor1)
```

```python
tensor1 = torch.logspace(1, 3, 3, base=2)
print(tensor1)
```

6.3.3 按数值填充张量

```python
tensor1 = torch.zeros(2, 3)
print(tensor1)
print(tensor1.dtype)
```

```python
tensor2 = torch.ones_like(tensor1)
print(tensor2)
```

```python
tensor1 = torch.full((3,2), 6.5)
print(tensor1)
print(tensor1.dtype)
```

```python
tensor2 = torch.empty_like(tensor1)
print(tensor2)
```

```python
tensor1 = torch.eye(3)
print(tensor1)
```

```python
tensor1 = torch.eye(3, 5)
print(tensor1)
```

6.3.4 随机张量创建

```python
tensor1 = torch.rand(2, 3)
print(tensor1)
```

```python
tensor2 = torch.rand_like(tensor1)
print(tensor2)
```

```python
tensor1 = torch.randint(low=0, high=100, size=(2, 3))
print(tensor1)
```

```python
tensor2 = torch.randint_like(tensor1, low=10, high=20)
print(tensor2)
```

```python
tensor1 = torch.randn(4, 2)
print(tensor1)
```

```python
tensor2 = torch.randn_like(tensor1)
print(tensor2)
```

```python
tensor1 = torch.normal(5, 2, tensor2.shape)
print(tensor1)
```

```python
# 随机重新排列
tensor1 = torch.randperm(10)
print(tensor1)
```

```python
# 查看和指定随机数种子
print(torch.random.initial_seed())
```
