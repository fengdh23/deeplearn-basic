```python
import torch
```

```python
# 1. 四则运算
tensor1 = torch.randint(low=0, high=10, size=(2,3))
print(tensor1)
```

```python
print(tensor1 + 10)
```

```python
print(tensor1.add(10))
```

```python
print(tensor1)
```

```python
print(tensor1.add_(10))
print(tensor1)
```

```python
print(tensor1.sub_(10))
```

```python
# 幂函数
print(tensor1 ** 2)
```

```python
print(tensor1.pow(2))
print(tensor1)
```

```python
print(tensor1.pow_(2))
print(tensor1)
```

```python
print(tensor1.sqrt())
```

```python
print(tensor1)
```

```python
tensor1 = tensor1.float()
print(tensor1.sqrt_())
print(tensor1)
```

```python
tensor1 = torch.tensor([1.0, 2, 3])
print(tensor1)
print(2.7182818 ** tensor1)
```

```python
print(tensor1.exp())
```

```python
print(tensor1)
print(tensor1.exp_())
print(tensor1)
```

```python
# 计算张量的哈达玛积
tensor1 = torch.tensor([[1, 2], [3, 4]])
tensor2 = torch.tensor([[1, 2], [3, 4]])
print(tensor1 + tensor2)
print(tensor1 * tensor2)
```

```python
print(tensor1.mul(tensor2))
```

```python
# 张量乘法
tensor1 = torch.tensor([[1, 2, 3], [4, 5, 6]])
tensor2 = torch.tensor([[1, 2], [3, 4], [5, 6]])
print(tensor1 @ tensor2)
```

```python
print(tensor1.matmul(tensor2))
```

```python
print(tensor1.mm(tensor2))
```

```python
tensor1 = torch.tensor([[[1, 2, 3], [4, 5, 6]], [[6, 5, 4], [3, 2, 1]]])
tensor2 = torch.tensor([[[1, 2], [3, 4], [5, 6]], [[6, 5], [4, 3], [2, 1]]])
print(tensor1 @ tensor2)
print(tensor1.matmul(tensor2))
```

```python
# print(tensor1.mm(tensor2))
tensor1 = torch.randn(4, 2, 2, 3)
tensor2 = torch.randn(1, 1, 3, 5)
tensor3 = tensor1 @ tensor2
print(tensor3.shape)
```

```python
# 验证内存的共享和节省内存
X = torch.randint(1, 10, (3, 2, 4))
print(X)
print(id(X))
```

```python
# X = X + 10
X += 10
print(X)
print(id(X))
```

```python
X = torch.randint(1, 10, (3, 2, 4))
Y = torch.randint(1, 10, (3, 4, 1))
print(X.shape)
print(id(X))
print(X)
print(Y)
print(X @ Y)
# X = X @ Y
# X @= Y
X[:] = X @ Y
print(X.shape)
print(X)
print(id(X))
```
