```python
import torch
```

```python
tensor1 = torch.randint(1, 10, (3, 5, 4))
print(tensor1)
```

```python
# 1. 简单索引
print(tensor1[2, 1, 3])
```

```python
print(tensor1[2, 3])
```

```python
print(tensor1[1])
```

```python
print(tensor1[:, 3])
```

```python
# 2. 范围索引
print(tensor1[1:])
```

```python
print(tensor1[-1:, 1:4, 0:3])
```

```python
print(tensor1[-1:, 1:4:2, 0:3:2])
```

```python
# print(tensor1[-1:, 4:1:-2, 0:3:2])
print(tensor1[-1:, -4:-1:2, 0:3:2])
```

```python
# 3. 列表索引
print(tensor1[[0, 1], [3, 2]])
```

```python
print(tensor1[[0, 1], [3, 2], [2, 3]])
```

```python
print(tensor1[[[0], [1]], [1, 2]])
```

```python
# 4. 布尔索引
# 4.1 选取符合条件的行：首元素大于5
mask = tensor1[:, :, 0] > 5
print(mask)
```

```python
print(tensor1[mask])
```

```python
# 4.2 选取符合条件的列：第二个元素大于5
mask = tensor1[:, 1, :] > 5
print(mask)
```

```python
# print(tensor1[mask])
tensor2 = tensor1.mT
print(tensor2)
```

```python
# 将mask应用到tensor2上，形状匹配前两个维度
print(tensor2[mask].mT)
```

```python
# 4.3 选取符合条件的矩阵: 元素（1,2）>5
mask = tensor1[:, 1, 2] > 5
print(mask)
```

```python
print(tensor1[mask])
```

```python
mask = tensor1 > 5
print(mask)
```

```python
print(tensor1[mask])
```
